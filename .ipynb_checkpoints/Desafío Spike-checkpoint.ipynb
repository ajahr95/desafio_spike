### Funciones

```python
def valid_period(dataframe, period_name, print_column=False):
    
    ''' Función para imprimir la primer y última fecha de valores válidos (que no sean NaN) 
        del dataframe respecto a cada una de sus columnas'''
    
    common_first = []
    common_last = []
    
    for column in dataframe:
        
        first = dataframe[period_name][pd.Series.first_valid_index(dataframe[column])]
        last = dataframe[period_name][pd.Series.last_valid_index(dataframe[column])]
        
        common_first.append(first)
        common_last.append(last)

        #se imprime caso por caso solo si se quiere ver
        if print_column:
            
            print(f'{column} - Index Válidos [{last}, {first}]')
            print()
        
    common_first.sort()
    common_last.sort()
    
    print(f'Primer válido común: {common_first[0]}')
    print(f'Último válido común: {common_last[-1]}')
```

```python
def period_to_datetime(period):
    
    ''' Función que recibe un string de periodo de la forma yyyy-MM-dd hh:mm:ss y retorna la misma fecha
        en formato Timestamp'''
    
    year = period[0:4]
    month = period[5:7]
    
    return pd.to_datetime(str(year) + ' , ' + str(month))
```

```python
def replace_values_types(value, column_type):
    
    ''' Función que revisa si existen strings no numéricos y los remplaza por NaN'''
    
    data_type = type(value).__name__
     
    if pd.isna(value) or data_type == 'Timestamp':
        
        return value
    
    else:
        
        if data_type == 'float' or data_type == 'int':
            
            return value
        
        elif data_type == 'str':
        
            new_value = value.replace('.', '')
            new_value = new_value.replace(',','')
        
            if new_value.isnumeric():
                
                
                if column_type[:3] == 'Ima':
                    
                    
                    if int(new_value[:3]) < 150:
                        
                        
                        new_value = new_value[:3] + '.' + new_value[3:]
                    
                    else:
                        
                        new_value = new_value[:2] + '.' + new_value[2:]
                        
                    
                elif column_type[:3] == 'PIB':
                    
                    
                    if len(column_type) == 3:
                        
                        new_value = new_value[:5] + '.' + new_value[5:]
                        
                    
                    else:
                        
                        new_value = new_value[:4] + '.' + new_value[4:]
                        
                        
                elif column_type[:3] == 'Imp' or column_type[:3] == 'Der':
                    
                    new_value = new_value[:4] + '.' + new_value[4:]
                    
                    
                elif column_type[:3] == 'Pre':
                    
                    new_value = new_value[:3] + '.' + new_value[3:]
                    
                    
                elif column_type[:3] == 'Tip':
                    
                    if column_type[-1] == 'o':
                        
                        new_value = new_value[:3] + '.' + new_value[3:]
                    
                    else:
                        
                        if int(new_value[:3]) < 150:

                            new_value = new_value[:3] + '.' + new_value[3:]


                        else:

                            new_value = new_value[:2] + '.' + new_value[2:]
                        
                elif column_type[:3] == 'Ocu':
                    
                    
                    if len(column_type) == 8:
                        
                        new_value = new_value[:4] + '.' + new_value[4:]
                        
                    
                    else:
                        
                        new_value = new_value[:3] + '.' + new_value[3:]
                        
                
                elif column_type[:3] == 'Ind':
                    
                    if int(new_value[:3]) < 200:
                        
                        new_value = new_value[:3] + '.' + new_value[3:]
                        
                    
                    else:
                        
                        new_value = new_value[:2] + '.' + new_value[2:]
                        
                        
                return float(new_value)
            
            else:
                
                return float('NaN')
                
         
                
        else:
            
            return value 
```

```python
def fix_dataframe(dataframe):
    
    ''' Función que transforma todos los valores de las columnas no numéricas de un dataframe 
        a NaN o a float, con los decimales establecidos en los supuestos'''
    
    new_dataframe = dataframe
    
    for column in dataframe.columns:
            
        new_dataframe[column] = new_dataframe[column].apply(replace_values_types, args=[column])
        
    return new_dataframe
```

#### Problema Puntuación

Se puede ver que la base de datos viene casi en su totalidad como números en formato string, NaN Y floats. El problema con los números como strings es que gran parte tiene puntos cada 3 dígitos, lo que no permite inferir si es que provienen de decimales, de separadores de miles o de ninguno de los anteriores. Ante eso, se debieron tomar supuestos (lo ideal en un caso real habría sido consultar al proveedor de la base de datos o al cliente).

`Supuesto Imacec`: dado los valores nacionales, se va a asumir que este debe ser menor a 150, por lo tanto se va a adaptar el decimal al punto de que cada valor lo cumpla.

`Supuesto PIB`: en la columna 'PIB_Industria_Manufacturera' hay un valor 1358.56, siendo este el único distinto que no tiene el punto cada 3 dígitos, por ende se va a asumir que hay una unidad detrás de millones o miles de millones. Ante esto el decimal se considerará al cuarto dígito. En el caso del PIB total (posiblemente) se considerará luego del quinto dígito.

`Supuesto IVA y DI`: al ser probablemnte valores altos también como el PIB y estar entre medio de sus columnas, se asumirá el decimal posterior al cuarto dígito también.

`Supuesto Precio gasolina y kerosene m3`: sus valores debiesen ser menores a 1000 dólares, por lo que su decimal se encontrará luego del tercer dígito.

`Supuesto Tipo de Cambio`: el TMC es un tipo de cambio que es menor a 150 normalmente, por lo que realizará la misma regla que Imacec. Para el tipo de cambio de valor observado se dejará el decimal posterior al tercer dígito.

`Supuesto Ocupación`: se va a asumir que la unidad son miles por lo que a la columna Ocupación, se le asignará la separación decimal luego del cuarto dígito, mientras que a los de ocupación por sector luego del tercer dígito (obviamente esto debiese variar por sector pero al menos permite generar una escala entre sus valores con un margen de error asociado).

`Supuesto índice`: dado la columna con nombre 'Indice_de_tipo_de_cambio_real___TCR_promedio_1986_100' se entiende que los índices rodean los valores 100, por lo que se tomará el mismo supusto que Imacec y se dará el decimal luego del segundo o tercer dígito solo que dependiendo de si es menor de 200 (no 150).

Con estos supuestos se procede a modificar los valores mal asignados en el dataframe con la función `fix_dataframe(central_bank)`.

```python
central_bank = fix_dataframe(central_bank)
```

```python
def time_series(dataframe ,columns, init_date, end_date, dataframe_type):
    
    ''' Función que recibe un dataframe, nombres de columnas, una fecha de inicio, de término y un datafram_type
        que puede ser Rain o PIB y retornará una serie de tiempo de esas variables en ese periodo'''
    
    color = sns.color_palette("tab10")

    color_counter = 0
    
    # agregamos un contador que confirme que se cumplan todas las condiciones de entrada
    condition = len(columns)
    
    
    for column_name in columns:
    
        if column_name in dataframe.columns:

            init_date = pd.to_datetime(init_date)
            end_date = pd.to_datetime(end_date)

            if init_date in dataframe.Periodo.values and end_date in dataframe.Periodo.values:

                if init_date <= end_date:
                    
                    condition -= 1

                    new_dataframe = dataframe.loc[(dataframe.Periodo.values < end_date) &
                                                  (dataframe.Periodo.values > init_date),
                                                 ['Periodo', column_name]]
                                        
                    sns.lineplot(data= new_dataframe,
                                 x='Periodo',
                                 y=column_name,
                                color=color[color_counter],
                                label=column_name)
                    
                    color_counter += 1
                    
                    if color == len(columns):
                        
                        color_counter = 0
                    
                else:

                    print('La fecha de inicio no puede ser mayor a la fecha final.')
                    break

            else:

                print('Una de las fechas no existe en los datos.')
                break

        else:

            print('Nombre de región inexistente')
            break
            
    if condition == 0:
            
        plt.tight_layout()

        column_name0 = columns[0].replace('_', ' ')
        column_name1 = columns[1].replace('_', ' ')

        plt.legend(labels=[column_name0, column_name1])

        if dataframe_type == 'Rain':

            plt.title('Precipitaciones región '+ column_name0 + ' y ' + column_name1, fontsize=14)

            plt.xlabel('Years')
            plt.ylabel('Rainfall (mm)')


        else:

            plt.title(column_name0 + ' y ' + column_name1, fontsize=14)

            plt.xlabel('Date')
            plt.ylabel('PIB')

        plt.show()
```

```python
def multiple_time_series(dataframe, years, region):
    
    '''Función que imprime múltiples series de tiempo de precipitaciones,
       respecto a los años y la región entregada'''
    
    if region in dataframe.columns:

        # filtramos por los años entregados y ordenamos por fecha el dataframe retornado
        new_dataframe = dataframe.loc[dataframe['year'].isin(years)]
        new_dataframe = new_dataframe.sort_values('Periodo')

        # definimos la grilla
        grid = sns.FacetGrid(
            data=new_dataframe,
            palette='hls',
            hue='year',
            col='year',
            col_wrap=3,
            height=4,
            aspect = 1,
            margin_titles=False,
            sharex = False)

        #agregamos el mapeo por los atributos presentados
        grid.map(plt.plot, 'month', region)

        # ajustamos la visualización de los ejes
        grid.set_axis_labels('month', 'Rainfall (mm)')

        for ax in grid.axes.flat:
            for label in ax.get_xticklabels():
                label.set_rotation(90)

        grid.fig.tight_layout()

        # ajustamos la altura y tamaño del título
        grid.fig.subplots_adjust(top=0.85)
        region = region.replace('_',' ')

        grid.fig.suptitle(f'Series de tiempo por año de la región {region}', fontsize = 14)

        plt.show()
        
    else:
        
        print('Nombre de región inexistente')
```

```python
def scatter_plot(dataframe, columns ,init_date, end_date):
    
    init_date = pd.to_datetime(init_date)
    end_date = pd.to_datetime(end_date)

    color = sns.color_palette("tab10") 
    
    if columns[0] in dataframe.columns and columns[1] in dataframe.columns:


        if init_date in dataframe.Periodo.values and end_date in dataframe.Periodo.values:

            if init_date <= end_date:

                new_dataframe = dataframe.loc[(dataframe.Periodo.values < end_date) &
                              (dataframe.Periodo.values > init_date),
                             [columns[0], columns[1]]]

                sns.scatterplot(data=new_dataframe,
                                x=columns[0],
                                y=columns[1])


                plt.tight_layout()

                column_name0 = columns[0].replace('_', ' ')
                column_name1 = columns[1].replace('_', ' ')

                plt.title('Gráfico de dispersión de ' + column_name0 + ' y ' + column_name1, fontsize=14)

                plt.xlabel(column_name0)
                plt.ylabel(column_name1)

                plt.show()

            else:

                print('La fecha de inicio no puede ser mayor a la fecha final.')

        else:

            print('Una de las fechas no existe en los datos.')

    else:

        print('Nombre de una de las regiones es inexistente')
```

```python
def month_transformation(month_number):
    
    '''Función que recibe un número de mes y retorna su nombre en español'''
    
    return month_names[month_number.month - 1]
```

```python
def year_transformation(timestamp_object):
    
    '''Función que recibe un objeto timestamp y retorna su año como entero'''
    
    return timestamp_object.year
```

### Funciones

```python
def data_to_datetime(year, month):
    
    ''' Función que recibe un año y un mes y lo transforma a Timestamp'''
    
    month_translation = {'Ene': '1', 'Feb': '2', 'Mar': '3', 'Abr': '4', 'May': '5', 'Jun': '6',
                        'Jul': '7', 'Ago': '8', 'Sep': '9', 'Oct': '10', 'Nov': '11', 'Dic': '12' }
    
    translation = str(year) + ', ' + month_translation[month]
    
    return pd.to_datetime(translation)
```

```python
def milk_price_plot(dataframe, x, y):
    
    '''Función que gráfica una serie de tiempo de la leche respecto a su fecha (x) y su precio(y)'''
    
    sns.lineplot(data=milk_price,
            x=x,
            y=y)
    
    plt.tight_layout()
    
    plt.title('Precio de la leche (1979-2020)', fontsize=14)
    
    plt.xlabel('Fecha')
    plt.ylabel('Precio Leche')
            

    plt.show()
    
    
```

```python
def trimester(date):
    
    '''Función que recibe un fecha y retorna a qué trimestre del año pertenece'''
    
    month = date.month
    
    if 1 <= month < 4:
        
        return 1
    
    elif 4 <= month < 7:
    
        return 2
    
    elif 7 <= month < 10:
    
        return 3
    
    elif 10 <= month <= 12:
        
        return 4
        
        
```

```python
def month_number(date):
    
    '''Función que recibe una fecha y retorna el número de su mes'''
    
    return date.month
```

#### NaN

Con la función `valid_period(all_data, period_name)` se puede obtener el primer y último dato válido (libre de NaN) en común a todas las columnas. En otras palabras, si se quisiese trabajar con un dataframe libre de NaN, se podría considerar solo este periodo.

```python
# agregar tercer argumento True si se quiere ver el detalle
valid_period(all_data, 'fecha')
```

Respecto a los NaN hay al menos 4 cosas que se podrían hacer:

`Recortar el dataframe a solo el periodo válido (sin NaN)`: de acuerdo a la función `valid_period(all_data, 'fecha)`, considerando todas las variables, el periodo valido sería entre 2014-01-01 y 2020-04-01. Si bien esto permite el uso directo sobre el dataframe en el modelo, pasaríamos de tener 506 datos (la cantidad de precios de leche disponibles) a solo 76, reduciendo el set de entrenamiento y testeo en gran parte.

`Reemplazarlos por la media o mediana de cada columna`: esta medida tiene como atractivo que su implementación es relativamente directa, pero sus resultados no necesariamente serán los correctos. Si de 506 filas, al menos un 75% de sus columnas están rellenadas en más de un 80% con medias o medianas podría generar una predicción erronea debido a los datos que usó para entrenarse.

`Eliminar variables económicas`: ya que los indicadores económicos son los limitantes por sus NaN (las precipitaciones tienen 496 datos, casi lo mismo que el precio de la leche), se podrían eliminar los que tienen menos datos, ya que hacen más corto el periodo total de análisis. Esto tiene como contra que conseguir más datos (pocos en comparación a los 506 de la leche), se perderían variables predictivas en el modelo.

`Utilizar una técnica como KNN`: esta técnica encuestra las K muestras más cercanas al valor NaN y reemplaza su valor a través de la obtención de la media. Se puede implementar con la librería sklearn e impyuete. (fuente: [towardsdatascience](https://towardsdatascience.com/whats-the-best-way-to-handle-nan-values-62d50f738fc))

`Utilizar una técnica como MICE`: lo que hace esta técnica es generar una regresión lineal en cada columna que contenga datos NaN y llenar sus valores dependiendo lo que entrege la regresión. (fuente: [towardsdatascience](https://towardsdatascience.com/whats-the-best-way-to-handle-nan-values-62d50f738fc))

Esta decisión también va a depender del modelo seleccionado.

## Modelo

### Funciones

```python
def to_ordinal(timestamp):
    
    ''' Función que transforma una fecha timestamp a ordinal'''
    
    return timestamp.toordinal()
```

```python
def milk_regression(milk_price, year, month, test_size):
    
    ''' Función que predice el precio de la leche con una regresión lineal respecto a un mes, un año 
        y la proporción entre el set de testeo y entrenamiento'''
    
    # se genera la regresión
    milk_reg = LinearRegression()
    
    # se establece la columna precio para ser utilizada como la variable a predecir
    price = milk_price['Precio_leche']

    # se transforma la fecha a ordinal para ser aplicada en el modelo solo si no se ha cambiado antes
    if type(milk_price['fecha'][0]).__name__ == 'Timestamp':
    
        milk_price['fecha'] = milk_price['fecha'].apply(to_ordinal)
    
    # se genera un set de entrenamiento de variables predictivas solo con la fecha
    train_data = milk_price.drop(['Anio', 'Mes', 'Precio_leche'], axis=1)
    
    # se separan los sets de entrenamiento y test en proporcion 0.9 a 0.1
    x_train, x_test, y_train, y_test = train_test_split(train_data.values, price.values, test_size=test_size, random_state=42)
    
    # se realiza el ajuste de la regresión
    milk_reg.fit(x_train, y_train)
    
    print(f'El score fue de: {str(round(milk_reg.score(x_test, y_test),2))}')
    
    print()
    
    #se genera la fecha a predecir como número ordinal
    new_date = pd.to_datetime(f'{year}, {month}')
    new_date = to_ordinal(new_date)
    price = milk_reg.predict(np.array([new_date]).reshape(1, 1))
    
    print(f'Precio de la leche del {year}-{month}-01 será: ' + str(round(price[0],2)))
```

#### Entrenamiento del modelo (Caso Regresión Lineal - No considera todos los dataframes)

Este modelo es solamente para ejemplificar un uso, ya que solo considera la fecha y el precio histórico del precio de la leche y a través de una regresión lineal entrega una predicción de como va a ser el precio. Para este caso, este es un modelo bastante pobre ya que toma en cuenta todo el dataset disponible de precipitaciones y de indicadores económicos para generar un resultado más preciso.

```python
#precio de la leche para 2021-03-01 con 10% de test de entrenamiento
milk_regression(milk_price, 2021, 3, 0.1)
```

El score R2 es 0.94 lo que indica una alto ajuste entre la regresión y los datos (esta no debe ser la única métrica a mirar, se pueden verificar errores también por un posible sesgo de valor). Se puede probar con distintas proporciones de set de entrenamiento y test (los resultados son similares).

#### Entrenamiento del modelo (ARIMA)

Este modelo busca especialmente predecir comportamientos de series de tiempo, pero a diferencia de la regresión lineal, condisera errores aleatorios y comporamientos cíclicos o estacionales para generar una predicción de manera estocástica. Para esto la serie debe pasar antes por un proceso de transformación a serie estacionaria y además, sedeben entregar los valores p (número de observaciones lags), d (grado de diferenciación), q (tamaño de la ventama movil promedio), lo que requiere un pre-procesamiento realizar el modelo. (fuente: [towardsdatascience](https://towardsdatascience.com/time-series-forecasting-predicting-stock-prices-using-an-arima-model-2e3b3080bd70))

#### Datos adicionales

Algunos datos adicionales que pudieron ser relevantes para el modelo:

`Temperatura y Humedad promedio`: la temperatura y humedad puede afectar a las condiciones del pasto o quizas a la misma capacidad de la vaca para dar leche. Por ende esta variable podría entregar valor ya que podría generar más gastos al tener que comprar más alimento o menos producción de leche.

`Mortalidad mensual de las vacas`: la mortalidad de las vacas puede tar un patrón de si esto ha ido en aumento o disminución a lo largo de tiempo (por más vacunas, remedios o condiciones climáticas), lo que aumenta el costo de producción y tambíen en consecuencia el precio.

`Natalidad mensual de las vacas`: la natalidad indicaría también si es que hay tendencia o estacionalidades respecto a los nacimientos a lo largo del tiempo. Entre más nacimientos, más producción de leche se tiene, por ende afecta al precio.

`m2 de pastoreo promedio por vaca`: una condición como esta puede afectar al estado físico de la vaca y a cuánta cantidad de leche producirá.

`Promedio de ordeñadas mensuales`: esto podría indicar si hay meses que se ordeña más a las vacas que otros, indicando estacionalidad y/o tendencia.

`Crecimiento mensual de industrias lacteas (chocolates, helados, etc)`: este indicador económico diría si es que se está demandando más la leche, lo que por efecto podría traer un aumento de precio por parte del productor.

`Promedio mensual del costo de alimento de las vacas (a veces se les da soya u otros derivados)`: para los casos en que se les da alimento adicional (no solo pastoreo), esto sería un costo adicional por lo que aumentaría el precio. Respecto a esto mismo se podrían considerar costos básicos como luz, agua, gas y gastos varios por el establecimiento.

`Inflación`: este indicador ayudaría a dar referencia respecto a un factor que obligadamente afecta al precio.

`Cantidad de festividades "chocolateras mensuales`:  esto sería un indicador de cuántas festividades en que se consume chocolate hay en el mes (navidad, pascua, etc), ya que podría afectar a la demanda y por ende el precio.
